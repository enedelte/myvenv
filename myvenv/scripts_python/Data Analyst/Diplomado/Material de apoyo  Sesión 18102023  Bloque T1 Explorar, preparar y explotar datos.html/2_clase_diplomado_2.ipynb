## Media y desviación estándar

Volvamos a nuestra visualización de la distribución de edades de los músicos.

* ¿Cuál dirías que es la edad típica de un músico en Melody Metropolis?
* ¿La mayoría de los músicos tienen aproximadamente esta edad o hay muchos músicos de diferentes edades?

Para responder estas preguntas de manera más específica, deberíamos tomar algunas medidas de nuestra variable.

* **La media**, también llamada promedio, describe el centro de una distribución numérica sumando todos los valores y dividiendo por el recuento.
* **La desviación estándar** describe la dispersión de valores en una distribución numérica en relación con la media. Se calcula encontrando la distancia promedio al cuadrado desde cada punto de datos a la media y aplicando la raíz cuadrada al resultado.

La edad media de los músicos es de 40,6 años y la desviación estándar es de unos 9,3 años. Podríamos interpretar esta desviación estándar como una variabilidad moderada en la edad.

Si la desviación estándar hubiera sido de 1 año, podríamos decir que apenas hay variabilidad en la edad:

4 edades desv.png

La media y la desviación estándar son opciones comunes, especialmente para distribuciones normales. Sin embargo, la media y la desviación estándar **no siempre son las mejores medidas** para describir una distribución.

## distribuciones asimétricas

Inspeccionamos la variable de ingresos.

5 ingresos.png

1. Hay bastantes músicos con mayores ingresos que están creando una cola más larga en el lado derecho.
2. La media indica que el ingreso típico es \$34,795. Este valor parece un poco alto ya que la mayoría de los ingresos parecen estar entre \$15.000 y \$40.000.

**La distribución del ingreso está sesgada**. Una distribución asimétrica es asimétrica con un cambio pronunciado de frecuencia en un lado y un cambio de frecuencia más plano y posterior en el otro.

Cuando los datos están sesgados, **la media puede no ser la mejor medida de una observación típica**.



## Mediana y IQR

Una alternativa sería encontrar el valor medio cuando todos los valores están ordenados de menor a mayor. **Este valor se llama mediana**, pero también se conoce como percentil 50 o segundo cuartil (Q2).

6 mediana.png

Estos datos abarcan 22 valores, que van del 6 al 28. Podríamos usar esto como nuestra medida de dispersión.

Una mejor medida podría ser el rango intercuartil (IQR). Un cuartil es simplemente un marcador de una cuarta parte (25%) de los datos.

* El primer cuartil marca el 25% (Q1 = 10).
* El segundo cuartil marca el 50% (Q2 = 13, la mediana)
* El tercer cuartil marca el 75% (Q3 = 22)

El IQR es la diferencia entre Q3 y Q1 (22-10 = 12), y marca el rango solo para el 50% medio de los datos.

7 rango inter.png

 Debido a que la **mediana y el IQR NO están muy influenciados por los valores extremos**, decimos que son robustos. Las estadísticas robustas suelen ser una mejor opción para medir el centro y la dispersión de una distribución sesgada o que tiene valores atípicos.

# **Análisis de datos exploratorios: estadísticas resumidas**

Las estadísticas resumidas son un componente importante del análisis de datos exploratorios (EDA) porque permiten a un analista de datos condensar una gran cantidad de información en un pequeño conjunto de números que pueden interpretarse fácilmente. Para decidir qué tipo de estadística resumida utilizar, es importante considerar dos cosas:

* Los datos (¿son cuantitativos o categóricos?)



## Variables cuantitativas
Al resumir variables cuantitativas, a menudo queremos describir la ubicación central y la distribución.



### tendencia central
Se utiliza a menudo para comunicar el valor "típico" de una variable. Recuerde que existen diferentes formas de calcular la ubicación central:

* Media: También llamada “promedio”; Se calcula como la suma de todos los valores dividida por el número de valores.
* Mediana: el valor medio de la variable cuando se ordena.
* Moda: El valor más frecuente de la variable.
* Media recortada: la media excluyendo el x por ciento de los puntos de datos más bajos y más altos.

La elección de una estadística resumida adecuada para la tendencia central a veces requiere técnicas de visualización de datos junto con conocimiento del dominio. Por ejemplo, supongamos que queremos saber el precio típico de un automóvil en nuestro conjunto de datos. ### Dispersión

La extensión o dispersión describe la variabilidad dentro de una característica. Esto es importante porque proporciona contexto para las medidas de ubicación central.

Al igual que las medidas de tendencia central, existen algunos valores que pueden describir la dispersión:
* Rango: La diferencia entre los valores máximo y mínimo de una variable.
* Rango Intercuartil (IQR): La diferencia entre los valores del percentil 75 y 25.
* Varianza: El promedio de la distancia al cuadrado desde cada punto de datos hasta la media.
* Desviación estándar (SD): La raíz cuadrada de la varianza.
* Desviación media absoluta (MAD): el valor medio absoluto de la distancia entre cada punto de datos y la media.

Elegir la medida de dispersión más adecuada es muy parecido a elegir una medida de tendencia central, en el sentido de que debemos considerar los datos de manera integral.

* Rango: Rs. 9970001
* RIQ: Rs. 420001

* Desviación estándar: Rs. 806253.40

## Tendencia central de datos cuantitativos

Para las variables cuantitativas, a menudo queremos describir la tendencia central o el valor "típico" de una variable. Por ejemplo, ¿cuál es el coste típico del alquiler en la ciudad de Nueva York?
Hay varias medidas comunes de tendencia central:

* Media: El valor promedio de la variable, calculado como la suma de todos los valores dividida por el número de valores.
* Mediana: el valor medio de la variable cuando se ordena.
* Moda: El valor más frecuente de la variable.
* Media recortada: la media excluyendo el x porciento de los puntos de datos más bajos y más altos.

Para nuestro DataFrame de alquileres con una columna denominada alquiler que contiene precios de alquiler, podemos calcular las estadísticas de tendencia central enumeradas anteriormente de la siguiente manera:

```python
from scipy.stats import trim_mean

# Mean
print(rentals.rent.mean())

# Median
print(rentals['rent'].median())

# Mode
print(rentals.rent.mode())

# Trimmed mean
print(trim_mean(rentals.rent, proportiontocut=0.1))  # delimita el 10% de los valores extremos
```

## Dispersión de datos cuantitativos

La dispersión de una variable cuantitativa describe la cantidad de variabilidad. Esto es importante porque proporciona contexto para las medidas de tendencia central. Por ejemplo, si hay mucha variabilidad en los precios de alquiler de la ciudad de Nueva York, podemos estar menos seguros de que el precio medio o mediano sea representativo del alquiler típico.

Hay varias medidas comunes de propagación:
* Rango: La diferencia entre los valores máximo y mínimo de una variable.
* Rango intercuartil (IQR): La diferencia entre los valores del percentil 75 y 25.
* Varianza: El promedio de la distancia al cuadrado desde cada punto de datos hasta la media.
* Desviación estándar (SD): La raíz cuadrada de la varianza.
* Desviación media absoluta (MAD): el valor medio absoluto de la distancia entre cada punto de datos y la media.

```python
#!pip install scipy
# Range
print(rentals.rent.max() - rentals.rent.min())

# Interquartile range
print(rentals.rent.quantile(0.75) - rentals.rent.quantile(0.25))

from scipy.stats import iqr
print(iqr(rentals.rent))  # alternative way

# Variance
print(rentals.rent.var())

# Standard deviation
print(rentals.rent.std())

# Mean absolute deviation
#print(rentals.rent.mad())
print((rentals.rent - rentals.rent.mean()).abs().mean())
```

## Conteo de valores para datos categóricos

## Proporciones de valores para datos categóricos

Podemos calcular la proporción de cada categoría dividiendo su recuento por el número total de valores de esa variable:
